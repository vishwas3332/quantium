```python
import pandas as pd
```

```python
all_files = ['data/daily_sales_data_0.csv','data/daily_sales_data_1.csv','data/daily_sales_data_2.csv']
```

```python
df_list=[]
```

```python
for filename in all_files:
    df = pd.read_csv(filename)
    df_list.append(df)
```

```python
mergerd_df = pd.concat(df_list,ignore_index=True)
mergerd_df
```

```python
# mergerd_df.to_csv("data/mergerd.csv",index=False)
```

```python
df_final = mergerd_df.loc[mergerd_df['product'] == 'pink morsel'].copy()

df_final['price'] = df_final['price'].replace(r'[\$,]', '', regex=True).astype(float)

df_final['Sales'] = df_final['quantity'] * df_final['price']
```

```python
df_final = df_final[['Sales', 'date', 'region']]
```

```python
df_final.to_csv("formatted_data.csv")
```
