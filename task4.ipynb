```python
from dash import Dash, html, dcc, Input, Output, callback
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('formatted_data.csv')
```

```python
fig = px.line(df,x='date',y='Sales',color='region')
```

```python
app = Dash()
app.layout = html.Div(children=[
    html.H1(children='Sales of Pink Morsel',style={"textAlign": "center"}),
    dcc.RadioItems(id='region-selector',options=df.region.unique(), value=df.region.unique()[0]),
    
    dcc.Graph(id='line-chart')
], style={"backgroundColor":"white"})


@app.callback(
    Output('line-chart','figure'),
    Input('region-selector','value')
)
def update_output(region):
    filtered_df = df[df.region == region]

    filtered_df = filtered_df.sort_values(by='date')

    fig = px.line(filtered_df,x='date',y='Sales',title=f'Sales Trends in {region}')

    fig.update_layout(
        title={'x':.5,'xanchor':'center'}
    )

    return fig

if __name__ == '__main__':
    app.run(debug=True)
```
