```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd
```

```python
app = Dash()
```

```python
df = pd.read_csv('formatted_data.csv')
```

```python
fig = px.line(df,x='date',y='Sales',color='region')
```

```python
app.layout = html.Div(children=[
    html.H1(children='Sales of Pink Morsel',style={"textAlign": "center", "color": "#04FFFB", "padding": "20px"}),
    dcc.Graph(
        id='sales-graph',
        figure=fig
    )
])
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```
